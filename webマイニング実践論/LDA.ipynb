```python
from gensim.corpora.dictionary import Dictionary
from gensim.models import LdaModel
from collections import defaultdict
from pathlib import Path
import itertools
import os
```

```python
path = './text/'
files = []

for x in os.listdir(path):
    if os.path.isfile(path + x):
        files.append(x) 
```

```python
# トピック数の設定
NUM_TOPICS = 100
```

```python
## 正規表現ライブラリの読み込み
import re

#　前処理　不必要な記号を削除
def preprocessing(document):

        lines = document.splitlines()
        processed_line = []

        horizontal_count = 0

        for line in lines:

            #line =re.sub(r'[!~]', '', line) #半角記号を除去
            #line =re.sub(r'[︰＠]', '', line) #全角記号を除去
            line = re.sub('\u3000', '', line) # \ufeffを除去
            #line = re.sub('する', '', line) # \ufeffを除去
            processed_line.append(line)

        return ''.join(processed_line)
```

```python
from natto import MeCab
def morphological(document):
        word = []
        word_list = []
        #MeCabの形態素解析結果のフォーマット
        with MeCab('-F%f[0],%f[1],%f[6]') as mcb:
            for i, token in enumerate (mcb.parse(document, as_nodes=True)):
                features = token.feature.split(',')
                #名詞（一般）動詞（自立）、形容詞（自立）以外は除外
                if features[0] == '名詞' and features[1] == '一般'  and features[2] != '' and features[2] != 'ストーリー' and features[2] != '作品' and features[2] != '映画' and features[2] != 'シリーズ' and features[2] != '脚本':
                #if features[0] == '名詞'  and features[2] != '':
                    word_list.append(features[2])
                #if features[0] == '動詞' and features[2] != '' and features[1] != 'する'　and features[1] != 'いる' and features[1] != 'れる':
                    #word_list.append(features[2])  #動詞は無視
              
                if features[0] == '形容詞' and features[1] == '自立' and features[2] != '':
                    word_list.append(features[2])
                
                if i%10000==0:
                    print(i, end=" ")
            word.append(word_list)
        return word_list
```

```python
lists = []
error = []
for filen in files:
    try:
        code='utf-8'#標準コード
        file='./text/{}'.format(filen)
        with open(file, 'r', encoding=code) as f:
            row_documents= f.read()
        docs=preprocessing(row_documents)
        dict=morphological(docs)
        lists.append(dict)
    except:
        error.append(filen)
```

```python
dictionary = Dictionary(lists)
corpus = [dictionary.doc2bow(text) for text in lists]
lda = LdaModel(corpus=corpus, num_topics=NUM_TOPICS, id2word=dictionary)
```

```python
def preprocessing2(document):

        lines = document.splitlines()
        processed_line = []

        horizontal_count = 0

        for line in lines:

            #line =re.sub(r'[!~]', '', line) #半角記号を除去
            line =re.sub('.txt', '', line) #全角記号を除去
            line = re.sub('\u3000', '', line) # \ufeffを除去
            #line = re.sub('する', '', line) # \ufeffを除去
            processed_line.append(line)

        return ''.join(processed_line)
```
