```python
import urllib.request
import csv
import pandas as pd
import numpy as np

cgi = 'https://ta-umedu.biwako.shiga-u.ac.jp/2019_web1/data1.cgi'
id = 'your_number'

years = [i for i in range(1990, 2010)]

for year in years:
    url = '{}?mode=show&dummy_id={}&year={}'.format(url, id, year)
    response = urllib.request.urlopen(url)
    data = response.read().decode('SJIS')
    reader = csv.reader(data.strip().split('\n'))
    
    header = next(reader)
    
    English = 0
    Mathematics = 0
    Science = 0
    Geography = 0

    count = 0
    
    for row in reader:
        count += 1
        English += int(row[1])
        Mathematics += int(row[2])
        Science += int(row[3])
        Geography += int(row[4])

    print('{}年：English{:.2f}, Mathematics{:.2f}, Science{:.2f}, Geography and History{:.2f}'.format(year, English/count, Mathematics/count, Science/count, Geography/count))
```
